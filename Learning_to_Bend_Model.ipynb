```python
import pandas as pd
import numpy as np
```

```python
#create input data X

x=np.linspace(0,10,50)
x=x.reshape(50,1)
x
```

```python
hidden_layer=3
```

**S3. Initialize Model Parameters**

intialize the weigths and bias

```python
w1=np.random.uniform(-1,1,(1,3))
b1=np.zeros((1,3))

w2=np.random.uniform(-1,1,(3,1))
b2=np.zeros((1,1))
```

```python
def activation(z):
  return np.maximum(0, z)
```

```python
def activation_slope(z):
    return (z > 0).astype(float)
```

forward pass

step1: linear transformation
step2: activation function
step3:

```python
z1 = x @ w1 + b1
h = activation(z1)
y_hat = h @ w2 + b2
```

```python
y = x * 2 + np.random.randn(*x.shape)
error = y_hat - y
```

```python
dl_dy= 2*error/len(x)

#gradients of O/P layer
dl_dw1= h.T@ dl_dy
dl_dw2= np.sum(dl_dy, axis=0, keepdims= True)

#apply activation slope
dL_dh = dl_dy @ w2.T
dh_dz1= dL_dh*activation_slope(z1)
```

```python
dL_dW1 = x.T @ dh_dz1
dL_db1 = np.sum(dh_dz1, axis=0, keepdims=True)
```

```python
learning_rate = 0.01

w1 -= learning_rate * dL_dW1
b1 -= learning_rate * dL_db1
w2 -= learning_rate * dl_dw2
b2 -= learning_rate * np.sum(dl_dy, axis=0, keepdims= True)
```

```python
learning_rate = 0.01
epochs = 1000

for epoch in range(epochs):

    z1 = x @ w1 + b1
    h = activation(z1)
    y_hat = h @ w2 + b2

    error = y_hat - y
    loss = np.mean(error ** 2)

    dL_dy = 2 * error / len(x)
    dL_dW2 = h.T @ dL_dy
    dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

    dL_dh = dL_dy @ w2.T
    dh_dz1 = dL_dh * activation_slope(z1)
    dL_dW1 = x.T @ dh_dz1
    dL_db1 = np.sum(dh_dz1, axis=0, keepdims=True)

    w1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1
    w2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2

    if epoch % 100 == 0:
        print("Epoch:", epoch, "Loss:", loss)
```
