```python
import pandas as pd
import numpy as np

columns = [
    "Sex", "Length", "Diameter", "Height",
    "WholeWeight", "ShuckedWeight",
    "VisceraWeight", "ShellWeight", "Rings"
]

data = pd.read_csv("abalone.data", names=columns)
print("Number of rows:", len(data))
print("Column names:", data.columns)
print(data.head())
```

```python
y = data["Rings"].values + 1.5
```

```python
X = data[["Length", "Diameter", "WholeWeight"]].values
```

```python
# Shuffle data
np.random.seed(42)
indices = np.random.permutation(len(X))

X = X[indices]
y = y[indices]

split = int(0.8 * len(X))

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
# Compute mean and std from training only
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)

# Normalize
X_train = (X_train - mean) / std
X_test = (X_test - mean) / std
```

```python
def forward(X, w, b):
    y_hat = X.dot(w) + b

    # Print shapes once
    print("X shape:", X.shape)
    print("w shape:", w.shape)
    print("b shape:", np.array(b).shape)
    print("y_hat shape:", y_hat.shape)

    return y_hat
```

```python
d = X_train.shape[1]

w = np.random.randn(d) * 0.01
b = 0.0
```

```python
def mse(y, y_hat):
    loss = ((y - y_hat) ** 2).mean()
    return loss
```

```python
def grad_w(X, y, y_hat):
    N = len(y)
    dW = (-2 / N) * X.T.dot(y - y_hat)
    return dW

def grad_b(y, y_hat):
    N = len(y)
    db = (-2 / N) * np.sum(y - y_hat)
    return db
```

```python
print("Gradient w shape:", grad_w(X_train, y_train, forward(X_train, w, b)).shape)
print("Gradient b shape:", np.array(grad_b(y_train, forward(X_train, w, b))).shape)
```

```python
lr = 0.01
epochs = 200

# Initial expectation:
# Loss should decrease gradually, not instantly.

for epoch in range(epochs):

    # 1) forward
    y_hat = X_train.dot(w) + b

    # 2) loss
    loss = mse(y_train, y_hat)

    # 3) gradients
    dW = grad_w(X_train, y_train, y_hat)
    db = grad_b(y_train, y_hat)

    # 4) update
    w = w - lr * dW
    b = b - lr * db

    if epoch % 20 == 0:
        print("Epoch:", epoch, "Loss:", loss)
```

```python
y_test_hat = X_test.dot(w) + b
```

```python
test_mse = mse(y_test, y_test_hat)
print("Test MSE:", test_mse)
```

```python
test_mae = np.abs(y_test - y_test_hat).mean()
print("Test MAE:", test_mae)
```
