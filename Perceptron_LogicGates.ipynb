```python
class Perceptron:
    def __init__(self, name, learning_rate=0.1, epochs=20):
        self.name = name
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = [0.0, 0.0]
        self.bias = 0.0

    def predict(self, x):
        # Activation: Step Function
        total_input = (self.weights[0] * x[0] + self.weights[1] * x[1]) + self.bias
        return 1 if total_input >= 0 else 0

    def train(self, inputs, labels):
        for epoch in range(self.epochs):
            for x, label in zip(inputs, labels):
                prediction = self.predict(x)
                error = label - prediction
                # Update weights and bias
                self.weights[0] += self.lr * error * x[0]
                self.weights[1] += self.lr * error * x[1]
                self.bias += self.lr * error

    def test(self, inputs, labels):
        print(f"--- Results for {self.name} Gate ---")
        print(f"Final Weights: {self.weights}")
        print(f"Final Bias: {self.bias}")
        print("Predictions:")
        success = True
        for x, label in zip(inputs, labels):
            pred = self.predict(x)
            status = "Correct" if pred == label else "FAILED"
            if pred != label: success = False
            print(f"  Input: {x} -> Predicted: {pred}, Actual: {label} [{status}]")

        if not success:
            print(f"NOTE: The Perceptron could not solve {self.name} because it is not linearly separable.")
        print("-" * 30 + "\n")


inputs = [[0, 0], [0, 1], [1, 0], [1, 1]]


gates = {
    "AND":  [0, 0, 0, 1],
    "OR":   [0, 1, 1, 1],
    "NAND": [1, 1, 1, 0],
    "NOR":  [1, 0, 0, 0],
    "XOR":  [0, 1, 1, 0]
}


for gate_name, labels in gates.items():
    model = Perceptron(gate_name)
    model.train(inputs, labels)
    model.test(inputs, labels)
```
